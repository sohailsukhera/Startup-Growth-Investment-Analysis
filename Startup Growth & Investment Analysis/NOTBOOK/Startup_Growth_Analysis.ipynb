# Startup Growth & Investment Analysis
### Using Python (Pandas, NumPy, Matplotlib)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../DATA/startup_growth_investment_data.csv")
df.head()
```

 # Basic Info

```python
df.info()
df.describe()
```

# COLUMNS UNDERSTAND

```python
df.columns
```

# MISSING VALUES CHECK

```python
df.isnull().sum()
```

# DATA CLEAING 

```python
df.isnull().sum()
df.dropna(inplace=True)
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

# Top 10 Startups by Valuation

```python
df.sort_values(by="Valuation (USD)", ascending=False).head(10)
```

 # Top 10 Highest Investment Startups

```python
df.sort_values(by="Investment Amount (USD)", ascending=False).head(10)
```

# Average Growth Rate by Industry

```python
df.groupby("Industry")["Growth Rate (%)"].mean().sort_values(ascending=False)
```

 # Top Countries by Startup Count

```python
df["Country"].value_counts()
```

 # Investors vs Growth Relationship

```python
df.groupby("Number of Investors")["Growth Rate (%)"].mean()
```

# Visualization 
# Industry Comparison

```python
import matplotlib.pyplot as plt

df["Industry"].value_counts().plot(kind="bar")
plt.title("Startups by Industry")
```

 # TOP 10 VALUED STARTUPS CHART

```python
top = df.sort_values("Valuation (USD)", ascending=False).head(10)
top.plot(x="Startup Name", y="Valuation (USD)", kind="bar")
plt.title("Top 10 Valued Startups")
plt.show()
```

# Growth Distribution

```python
df["Growth Rate (%)"].plot(kind="hist", bins=20)
plt.title("Growth Rate Distribution")
plt.show()
```
